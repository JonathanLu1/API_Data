```python
import pandas as pd
import numpy as np 
import requests
from citipy import citipy 
import time 
from datetime import datetime 
##from citipy import citipy

# Import API key
import os
from dotenv import load_dotenv
load_dotenv()
weather_key = os.getenv("OpenWeather_API_KEY")
```

```python
lat = np.random.uniform(low=-100.000, high=100.000, size=2000)
lng = np.random.uniform(low=-160.000, high=160.000, size=2000)
lat_lngs = zip(lat, lng)
lat_lngs 
```

```python
coordinates = list(lat_lngs) 
coordinates 
```

```python
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name 
    
    if city not in cities:
        cities.append(city) 
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_key
```

```python
city_data = []
print("Beginning Data Retrieval     ")
print("-----------------------------")

record_count = 1
set_count = 1 

for i in range(len(cities)):

    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    city_url = url + "&q=" + cities[i] 
    
    

    
for i, city in enumerate(cities): 

    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    city_url = url + "&q=" + city.replace(" ","+")

    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count += 1 
    try:
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_description = str(city_weather["weather"][0]["description"]) 
        city_max_temp = city_weather["main"]["temp_max"] 
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"] 
    
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Current Description": city_description,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    except:
        print("City not found. Skipping...")
        pass

print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------") 
```

```python
len(city_data)
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
city_data_clean = ['City', 'Country', 'Lat', 'Lng', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed', 'Current Description']
city_data_df = city_data_df[city_data_clean] 
city_data_df.head(10)   
```
